When a match is declared a `Walkover`, these matches become null as if it was never scheduled at all. This is when a player pulls out of competition before it even begins, often times due to injury. I decided to remove all rows that had `Walkover` in the `Comment` column.

```python
df_atp = df_atp[df_atp["Comment"].str.contains("Walkover") == False]
df_atp
```

```python
#Check to see other values in the comment column
df_atp.Comment.value_counts()
```

I noticed a lot of the values can be consolidated. 
- One of the comments has a typo with "Rrtired". I know this is intended to say "Retired". I decide to consolidate the two words.
- Awarded doesn't mean anything different than completed, so I chose to replace "Awarded" with "Completed".
- Disqualified means one of the players was disqualified usually for unsporting conduct. These matches are still considered completed with a winner and loser. There are only 3 occurrences of this happening. I decided to replace "Disqualified" with "Completed.
- Sched is just short for scheduled. I don't see these being any different than completed either so I replaced them with "Completed".

```python
#Replacing Rrtired with Retired 
df_atp['Comment'].replace({"Rrtired": "Retired"}, inplace=True)
```

```python
#Replacing Awarded with Completed
df_atp['Comment'].replace({"Awarded": "Completed"}, inplace=True)
```

```python
#Replacing Disqualified with Completed
df_atp['Comment'].replace({"Disqualified": "Completed"}, inplace=True)
```

```python
#Replacing Sched with Completed
df_atp['Comment'].replace({"Sched": "Completed"}, inplace=True)
```

```python
#Confirming that Rrtired has now been combined with Retired
df_atp.Comment.value_counts()
```

For the WRank and LRank, I decided to look up the lowesting ranking in ATP tennis and found that there's rankings up to 1814, so I decided to replace the null values with that number. These values make up a very small percentage of the data and figured this will not have a big impact.

```python
df_atp['LRank'] = df_atp['LRank'].replace(np.nan, 1814)
df_atp['WRank'] = df_atp['WRank'].replace(np.nan, 1814)
```

The only other null values are the WPts and LPts, they make up a very small percentage of the data as well. This time I decided to replace the null value with 0. Also, these players might not have earned any points at this point so likely could have been 0 in the first place.

```python
df_atp['LPts'] = df_atp['LPts'].replace(np.nan, 0)
df_atp['WPts'] = df_atp['WPts'].replace(np.nan, 0)
```

When it comes to the sets in tennis. Some matches are best of 3 or best of 5. There are likely null values in some of those sets because the match ended before being required to play more than the necessary amount of sets. There are even null values for the first and second sets. This could be due to a player retiring before those sets even began. I decided the best way to handle the null values is to make them 0.

```python
df_atp['W1'] = df_atp['W1'].replace(np.nan, 0)
df_atp['L1'] = df_atp['L1'].replace(np.nan, 0)
df_atp['W2'] = df_atp['W2'].replace(np.nan, 0)
df_atp['L2'] = df_atp['L2'].replace(np.nan, 0)
df_atp['W3'] = df_atp['W3'].replace(np.nan, 0)
df_atp['L3'] = df_atp['L3'].replace(np.nan, 0)
df_atp['W4'] = df_atp['W4'].replace(np.nan, 0)
df_atp['L4'] = df_atp['L4'].replace(np.nan, 0)
df_atp['W5'] = df_atp['W5'].replace(np.nan, 0)
df_atp['L5'] = df_atp['L5'].replace(np.nan, 0)
```

Finally the Wsets and Lsets also have null values. I decided to do some research to see if I find some common information about the rows that have null values.

```python
df_atp[df_atp['Wsets'].isna()]
```

```python
df_atp[df_atp['Lsets'].isna()]
```

I noticed that all these Wsets and Lsets with null values mentioned `Retired` when I viewed the `Comment` column. This means the matches ended prematurely and also by looking at each of the columns, I can tell not one of these matches even completed 1 set. I decided just to turn these null values into 0's.

```python
df_atp['Wsets'] = df_atp['Wsets'].replace(np.nan, 0)
df_atp['Lsets'] = df_atp['Lsets'].replace(np.nan, 0)
```
